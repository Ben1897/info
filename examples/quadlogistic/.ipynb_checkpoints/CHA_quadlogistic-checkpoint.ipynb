# Parameters

```python
# Plotting parameters
def get_filenames(folder, key, transred=True):
    netfile  = folder + 'eco_cmiknn_filtered6hr_' + key + 'RHTRgWS_results.pkl'
    if transred:
        filename = folder + 'info_eco_cmiknn_filtered6hr_' + key + 'RHTRgWS_results_tranred.pkl'
    else:
        filename = folder + 'info_eco_cmiknn_filtered6hr_' + key + 'RHTRgWS_results.pkl'
    datasrc = 'parents_eco_cmiknn_filtered6hr_' + key + 'RHTRgWS_results.pkl'
    
    return netfile, filename, datasrc
        
def get_title_labels(varnames):
    titles  = ['$-> ' + varn[1:-1] + '$' for varn in varnames]
    ylabels = ['$' + varn[1:-1] + '$ \n $[nats]$' for varn in varnames]
    return titles, ylabels

def get_varnames(varnames):
    varnames2 = []
    for varn in varnames:
        varnames2.append('$'+varn+'$')
    return varnames2

def reorder_data(data, order, index=[0]):
    if index == [0]:
        return data[order,:]
    elif index == [1]:
        return data[:,order]
    elif index == [0,1]:
        try:
            data2 = data[order,:,:]
            return data2[:,order,:]
        except:
            data2 = data[order,:]
            return data2[:,order]

# Parameters for the selected files
start, prtype = 0, 0
folder = "./networks/"

# info file
filename = "logistic_4var_anomaly_taumax5.pkl"
filenamet = "logistic_4var_anomaly_taumax5_transitive.pkl"

# net file
netfile = lambda T: 'logistic_cmiknn_parallel_30_T' + T + '_N4_full_shuffle_fullycoupled_ez02_results.pkl'

# Parameters for plotting
labels = [r'$\mathcal{J}$', r'$\mathcal{D}$', r'$\mathcal{T}$','IND']
labels2 = [r'$\mathcal{J}$ dominated', r'$\mathcal{P}$ dominated']
xlabel, ylabel = r'$\tau$', r'$[nats]$'
colors = ['blue', 'orange', 'black', 'red']
markers  = ['.', '.', '.', '.']
alphas = [1., 1., 1., 1]
figsize=(7,7)

# Parameters for network constructions
approach, k = 'knn', 5
taumax = 20
tauinter = 1.
taustart = 5
```

# Load the CHA results

```python
# Get the cit data
infos, tauranges, sizes = {}, {}, {}
infost, taurangest, sizest = {}, {}, {}
infos, tauranges, sizes, keysraw = load_cit_data4(folder+filename, prtype, start)
infost, taurangest, sizest, keysrawt = load_cit_data4(folder+filenamet, prtype, start)

# Sort the key
keys_temp = [int(key) for key in keysraw]
# keys = [str(key) for key in sorted(keys_temp)]
keys = ['200', '400', '600', '1000', '5000', '10000']

# Get the network
rds, datas, nets, varnsraws, lagfuncss, sigthress = {}, {}, {}, {}, {}, {}
for key in keys:
    data, net, varnsraw, lagfuncs, sigthres = load_net2(folder+netfile(key), taumax, approach, k)
    datas[key], nets[key] = data, net
    varnsraws[key], lagfuncss[key], sigthress[key] = varnsraw, lagfuncs, sigthres
    
# Get the variable names
varnss = {}
for key in keys:
#     varnss[key] = get_varnames(varnsraws[key])
    varnss[key] = get_varnames(['X_1', 'X_2', 'X_3', 'X_4'])

# Get the labels
nvars, titless, ylabels = {}, {}, {}
for key in keys:
    nvars[key] = len(varnss[key])
    titless[key], ylabels[key] = get_title_labels(varnss[key])

nvar, varns = len(varnss[key]), varnss[key]
```

# Plot the cardinalities of $\mathcal{J}$ and $\mathcal{D}$ with and without MIWTR

```python
fig, axes = plt.subplots(nrows=nvar,ncols=len(keys),figsize=(12,5),sharex=True,sharey=True)

for j in range(len(keys)):
    key = keys[j]
    dimsize1, taurange1, ylabel = sizes[key]['dimsize'], tauranges[key], ylabels[key]
    dimsize2, taurange2, ylabel = sizest[key]['dimsize'], taurangest[key], ylabels[key]
    for i in range(nvar):
        ax = axes[i,j]
        ax.plot(taurange1[taustart:], dimsize2[i,taustart:,1]-1, 'b-', label='$\mathcal{J}$ (MIWTR)')
        ax.plot(taurange2[taustart:], dimsize1[i,taustart:,1]-1, 'b--', label='$\mathcal{J}$ (original)')
        ax.plot(taurange1[taustart:], dimsize2[i,taustart:,2]-1, 'y-', label='$\mathcal{D}$ (MIWTR)')
        ax.plot(taurange2[taustart:], dimsize1[i,taustart:,2]-1, 'y--', label='$\mathcal{D}$ (original)')
        ax.set_xlim([-5, 60])

for j in range(len(keys)):
    axes[0,j].set_title(keys[j])
    axes[-1,j].set_xlabel(r'$\tau$')

for i in range(nvar):
    ax = axes[i,0]
    ax.set_ylabel(ylabel[i])
    axes[0,j].set_xticks([taustart,25,50])

plt.subplots_adjust(hspace=0, wspace=0)
axes[-1,3].legend(ncol=4, bbox_to_anchor=(2.5, -0.4), fontsize=15,frameon=False)
```

# Plot partial information decomposition on $\mathcal{D}$ over the partioning time lag $\tau$ and data length

```python
print 'PID for distant causal history...'
pidlabels = [r'$S_{\mathcal{D}}$', r'$R_{\mathcal{D}}$', r'$U_{self,\mathcal{D}}$', r'$U_{cross,\mathcal{D}}$']
# colors = ['dimgrey', 'blue', 'darkblue', 'silver']
colors = ['yellow', 'lawngreen', 'violet', 'lightgrey']

fig, axes = plt.subplots(nrows=nvar,ncols=len(keys),figsize=(12,5),sharex=True, sharey=True)
for j in range(len(keys)):
    key = keys[j]
    pidpit, taurange, ylabel = infost[key]['pidpit'], tauranges[key], ylabels[key]

    for i in range(nvar):
        ax = axes[i,j]
        base = np.zeros(taurange.size-taustart)
        for k in [2,1,3,0]:
            ax.fill_between(taurange[taustart:], base, base+pidpit[i,taustart:,k], 
                            color=colors[k], label=pidlabels[k])
            base += pidpit[i,taustart:,k]
        ax.set_xlim([-5, 60])

for i in range(nvar):
    ax = axes[i,0]
    ax.set_ylabel(ylabel[i])

for j in range(len(keys)):
    axes[0,j].set_title(keys[j])
    axes[-1,j].set_xlabel(r'$\tau$')
    axes[0,j].set_xticks([taustart,25,50])
axes[-1,-3].legend(ncol=4, bbox_to_anchor=(2, -0.4), fontsize=15,frameon=False)
plt.subplots_adjust(hspace=0, wspace=0)

fig, axes = plt.subplots(nrows=nvar,ncols=len(keys),figsize=(12,5),sharex=True, sharey=True)
for j in range(len(keys)):
    key = keys[j]
    taurange, ylabel = taurangest[key], ylabels[key]
    pidpit, pidpitt = infos[key]['pidpit'], infost[key]['pidpit']

    for i in range(nvar):
        ax = axes[i,j]
        for k in [2,1,3,0]:
            ax.axhline(0, linestyle='--', color='gray')
            ax.plot(taurange[taustart:], pidpit[i,taustart:,k]-pidpitt[i,taustart:,k], 
                    color=colors[k], label=pidlabels[k])
        ax.set_xlim([-5, 60])

for i in range(nvar):
    ax = axes[i,0]
    ax.set_ylabel(ylabel[i])

for j in range(len(keys)):
    axes[0,j].set_title(keys[j])
    axes[-1,j].set_xlabel(r'$\tau$')
    axes[0,j].set_xticks([taustart,25,50])
axes[-1,-3].legend(ncol=4, bbox_to_anchor=(2, -0.4), fontsize=15,frameon=False)
plt.subplots_adjust(hspace=0, wspace=0)
```
