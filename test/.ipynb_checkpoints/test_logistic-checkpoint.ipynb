```python
%reset
%load_ext autoreload
%autoreload 2

from logistic_network import Logistic
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Parameters for the relationship between variables
n    = 3
lag  = 2
adjM = np.array([[0, 1, 0], [1, 0, 1], [1, 0, 0]])
lagM = np.array([[0, lag, 0], [lag, 0, lag], [lag, 0, 0]])
# Parameters for simulation
nstep = 200
```

```python
# Parameters for noise parameters
# exclusive -- uniform
noiseType11 = 'exclusive'
noiseDist11 = 'uniform'
noisePara11 = [1, 0, 1]

# exclusive -- normal
noiseType12 = 'exclusive'
noiseDist12 = 'normal'
noisePara12 = [1, 0, 1]

# additive -- normal
noiseType21 = 'additive'
noiseDist21 = 'normal'
noisePara21 = [1, 0, 1]

# additive -- uniform
noiseType22 = 'additive'
noiseDist22 = 'uniform'
noisePara22 = [1, 0, 1]

# multiplicative -- uniform
noiseType31 = 'multiplicative'
noiseDist31 = 'uniform'
noisePara31 = [1, 0, 1]

# multiplicative -- normal
noiseType32 = 'multiplicative'
noiseDist32 = 'normal'
noisePara32 = [1, 0, 1]
```

```python
# Initialize the logistic equations
logistic11 = Logistic(n, adjM, lagM, noiseType11, noiseDist11, noisePara11)
logistic12 = Logistic(n, adjM, lagM, noiseType12, noiseDist12, noisePara12)
logistic21 = Logistic(n, adjM, lagM, noiseType21, noiseDist21, noisePara21)
logistic22 = Logistic(n, adjM, lagM, noiseType22, noiseDist22, noisePara22)
logistic31 = Logistic(n, adjM, lagM, noiseType31, noiseDist31, noisePara31)
logistic32 = Logistic(n, adjM, lagM, noiseType32, noiseDist32, noisePara32)
```

```python
# Simulate
results11 = logistic11.simulate(nstep)
results12 = logistic12.simulate(nstep)
results21 = logistic21.simulate(nstep)
results22 = logistic22.simulate(nstep)
results31 = logistic31.simulate(nstep)
results32 = logistic32.simulate(nstep)
```

```python
# Plot
t = range(nstep)
plt.figure(1)
plt.plot(t, results11[0, :], 'b')
plt.plot(t, results11[1, :], 'k')
plt.plot(t, results11[2, :], 'r')
plt.figure(2)
plt.plot(t, results12[0, :], 'b')
plt.plot(t, results12[1, :], 'k')
plt.plot(t, results12[2, :], 'r')
plt.figure(3)
plt.plot(t, results21[0, :], 'b')
plt.plot(t, results21[1, :], 'k')
plt.plot(t, results21[2, :], 'r')
plt.figure(4)
plt.plot(t, results22[0, :], 'b')
plt.plot(t, results22[1, :], 'k')
plt.plot(t, results22[2, :], 'r')
plt.figure(5)
plt.plot(t, results31[0, :], 'b')
plt.plot(t, results31[1, :], 'k')
plt.plot(t, results31[2, :], 'r')
plt.figure(6)
plt.plot(t, results32[0, :], 'b')
plt.plot(t, results32[1, :], 'k')
plt.plot(t, results32[2, :], 'r')
plt.show()
```
