```python
import numpy as np
import pandas as pd

import seaborn as sns
from scipy.stats import norm, multivariate_normal
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
from matplotlib import cm
import matplotlib.gridspec as gridspec

from info.core.ccm import ccm, extended_ccm
from info.plot import plot_ccm
from info.utils.others import corrcoefs, parse_SFP, aggregate, butter_filter
from info.utils.sst import conductSST_withlag
```

## Read the SFP data, filtered the data and conduct CCM and MI on it

```python
# Read the data
filepath = '../data/SFP2_AllData.mat'
times, data, header, doy = parse_SFP(filepath)
precip, rh, temp = data[:, 1], data[:, 2], data[:, 3]
solar, wind      = data[:, 4], data[:, 5]
```

```python
fs = 1./60.          # the sample rate
fc = 1./(60.*60.*6.) # the cutoff frequency (6hr)
N  = 5               # the order of the butterworth filter
precipf = butter_filter(precip, N, fs, fc, btype='high')
rhf     = butter_filter(rh, N, fs, fc, btype='high')
tempf   = butter_filter(temp, N, fs, fc, btype='high')
solarf  = butter_filter(solar, N, fs, fc, btype='high')
windf   = butter_filter(wind, N, fs, fc, btype='high')
```

```python
# Define the coefficients for CCM
tau_set = range(1, 21, 2)
E_set = range(2, 22, 2)
tsize = len(tau_set)
Esize = len(E_set)
# Settings for plot
plt.rcParams["axes.labelsize"] = 15
plt.rcParams["axes.titlesize"] = 15
plt.rcParams["xtick.labelsize"] = 15
plt.rcParams["ytick.labelsize"] = 15
```

### 1. A simple test on temperature and relative humidity

```python
lagmax, L = 300, 1000
# Analysis on the original data
lagset, rhoxmpy, rhoympx = extended_ccm(temp[:L], rh[:L], temp[:L], rh[:L], 2, lagmax, 1)
lagset, rhoympz, rhozmpy = extended_ccm(rh[:L], temp[:L], rh[:L], temp[:L], 2, lagmax, 1)
lagset, mirxy, mixy = conductSST_withlag(temp[:L], rh[:L], lagmax, returnTrue=True, approach='kde_cuda')
rhoxy = corrcoefs(temp[:L], rh[:L], lagset)
# Analysis on the filtered data
lagset, rhoxmpyf, rhoympxf = extended_ccm(tempf[:L], rhf[:L], tempf[:L], rhf[:L], 2, lagmax, 1)
lagset, rhoympzf, rhozmpyf = extended_ccm(rhf[:L], tempf[:L], rhf[:L], tempf[:L], 2, lagmax, 1)
lagset, mirxyf, mixyf = conductSST_withlag(tempf[:L], rhf[:L], lagmax, returnTrue=True, approach='kde_cuda')
rhoxyf = corrcoefs(tempf[:L], rhf[:L], lagset)
```

### 2.1. Tests on the accumulated and averaged data -- hourly data

```python
# Define
lagmax, L = 30, 1000
# Hourly data
interval = 60
prep_hr  = aggregate(precip, interval, method='accumulate')
rh_hr    = aggregate(rh, interval, method='average')
temp_hr  = aggregate(temp, interval, method='average')
solar_hr = aggregate(solar, interval, method='accumulate')
wind_hr  = aggregate(wind, interval, method='average')
# Hourly data (filtered)
fs = 1./(60.*60.)          # the sample rate
fc = 1./(60.*60.*6.) # the cutoff frequency (6hr)
# prepf_hr  = butter_filter(prep_hr, N, fs, fc, btype='high')
# rhf_hr    = butter_filter(rh_hr, N, fs, fc, btype='high')
# tempf_hr  = butter_filter(temp_hr, N, fs, fc, btype='high')
# solarf_hr = butter_filter(solar_hr, N, fs, fc, btype='high')
# windf_hr  = butter_filter(wind_hr, N, fs, fc, btype='high')
prepf_hr  = aggregate(precipf, interval, method='accumulate')
rhf_hr    = aggregate(rhf, interval, method='average')
tempf_hr  = aggregate(tempf, interval, method='average')
solarf_hr = aggregate(solarf, interval, method='accumulate')
windf_hr  = aggregate(windf, interval, method='average')
```

```python
x, y = temp_hr[:L], rh_hr[:L]
x_f, y_f = temp_hr[L:2*L], rh_hr[L:2*L]
xmpy, ympx = np.zeros([tsize, Esize]), np.zeros([tsize, Esize])
for i in range(tsize):
    tau = tau_set[i]
    for j in range(Esize):
        E = E_set[j]
        _, rho1 = ccm(x, y, x_f, y_f, E, tau)
        _, rho2 = ccm(y, x, y_f, x_f, E, tau)
        xmpy[i, j] = rho1
        ympx[i, j] = rho2
# Plot
extent = [E_set[0], E_set[-1], tau_set[-1], tau_set[0]]
xv, yv = np.meshgrid(tau_set, E_set, indexing='ij')
plot_ccm.plot_ccm_xy(xmpy, ympx, extent, 'E', 'tau')
plt.show()
```

```python
# Compute rho for different lengths of the time series
L = 1000
x, y = temp_hr[:L], rh_hr[:L]
x_f, y_f = temp_hr[:L], rh_hr[:L]
xmpy, ympx = [], []
rhoset = []
L_set = range(20, L, 20)
E, tau = 4, 2
for l in L_set:
    _, rho1 = ccm(x[:l], y[:l], x_f[:l], y_f[:l], E, tau)
    _, rho2 = ccm(y[:l], x[:l], y_f[:l], x_f[:l], E, tau)
    rhoset.append(np.corrcoef(x[:l], y[:l])[0,1])
    xmpy.append(rho1)
    ympx.append(rho2)

# Plot
plot_ccm.plot_ccm(L_set, xmpy, ympx, rhoset)
plt.show()
```

```python
# Plotting function
def plot_extended_ccm_xy_withmi(lagset, rhoxmpy, rhoympx, rhoxy, mirxy, mixy,
                                rhofxmpy, rhofympx, rhofxy, mirfxy, mifxy):
    """Plot the correlation coefficients and the CCM skills in terms of different lags for 2 nodes."""
    # Get the lags with the maximum skill
    maxympx, maxxmpy = lagset[np.argmax(rhoympx)], lagset[np.argmax(rhoxmpy)]
    maxfympx, maxfxmpy = lagset[np.argmax(rhofympx)], lagset[np.argmax(rhofxmpy)]

    # Get the lags with the correlation coefficients
    maxrhoxy = lagset[np.argmax(rhoxy)]
    maxrhofxy = lagset[np.argmax(rhofxy)]

    # Get the lags with the maximum skill
    maxmixy = lagset[np.argmax(mixy)]
    maxmifxy = lagset[np.argmax(mifxy)]
    
    xyt = mirxy==True
    xyf = mirxy==False
    xyft = mirfxy==True
    xyff = mirfxy==False

    # Plot
    plt.rcParams["figure.figsize"] = (16, 10)
    gs = gridspec.GridSpec(3, 2)
    gs.update(wspace=0.4, hspace=0.4)

    ax = plt.subplot(gs[0, 0])
    ax.plot(lagset, rhoxmpy, 'b', label='x xmp y')
    ax.plot(lagset, rhoympx, 'r', label='y xmp x')
    ax.set_xlabel('Lag')
    ax.set_ylabel('Correlation Coefficient')
    ax.set_ylim([-.5, 1.1])
    ax.set_title('Max. lag (x xmp y): %d; Max. lag(y xmp x): %d.' % (maxxmpy, maxympx))
    ax.legend(loc='lower right')

    ax = plt.subplot(gs[1, 0])
    ax.scatter(lagset[xyt], np.ones(xyt.sum()), color='red', label='MI SST True')
    ax.scatter(lagset[xyf], np.ones(xyf.sum()), color='blue', label='MI SST False')
    ax.plot(lagset, mixy, label='I value')
    ax.set_xlabel('Lag')
    ax.set_ylabel('I[x(t), y(t+Lag)]')
    ax.set_title('Max. lag: %d;' % maxmixy)
    ax.legend(loc='lower right')

    ax = plt.subplot(gs[2, 0])
    ax.plot(lagset, rhoxy, 'k', label='rho(x, y)')
    ax.set_xlabel('Lag')
    ax.set_ylabel('Rho[x(t), y(t+Lag)]')
    ax.set_title('Max. lag: %d;' % maxrhoxy)
    ax.set_ylim([-1.1, 1.1])
    
    ax = plt.subplot(gs[0, 1])
    ax.plot(lagset, rhofxmpy, 'b', label='x xmp y')
    ax.plot(lagset, rhofympx, 'r', label='y xmp x')
    ax.set_xlabel('Lag')
    ax.set_ylabel('Correlation Coefficient')
    ax.set_ylim([-.5, 1.1])
    ax.set_title('Max. lag (x xmp y): %d; Max. lag(y xmp x): %d.' % (maxfxmpy, maxfympx))
    ax.legend(loc='lower right')

    ax = plt.subplot(gs[1, 1])
    ax.scatter(lagset[xyft], np.ones(xyft.sum()), color='red', label='MI SST True')
    ax.scatter(lagset[xyff], np.ones(xyff.sum()), color='blue', label='MI SST False')
    ax.plot(lagset, mifxy, label='I value')
    ax.set_xlabel('Lag')
    ax.set_ylabel('I[x(t), y(t+Lag)]')
    ax.set_title('Max. lag: %d;' % maxmifxy)
    ax.legend(loc='lower right')

    ax = plt.subplot(gs[2, 1])
    ax.plot(lagset, rhofxy, 'k', label='rho(x, y)')
    ax.set_xlabel('Lag')
    ax.set_ylabel('Rho[x(t), y(t+Lag)]')
    ax.set_title('Max. lag: %d;' % maxrhofxy)
    ax.set_ylim([-1.1, 1.1])
```
