```python
import numpy as np
import pandas as pd

import seaborn as sns
from scipy.stats import norm, multivariate_normal
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
from matplotlib import cm
import matplotlib.gridspec as gridspec

from info.core.ccm import ccm, extended_ccm
from info.plot import plot_ccm
from info.utils.others import corrcoefs, parse_SFP, aggregate, butter_filter
from info.utils.sst import conductSST_withlag
```

## Read the SFP data, filtered the data and conduct CCM and MI on it

```python
# Read the data
filepath = '../data/SFP2_AllData.mat'
times, data, header, doy = parse_SFP(filepath)
prep, temp = data[:, 1], data[:, 3]
```

```python
fs = 1./60.          # the sample rate
fc = 1./(60.*60.*6.) # the cutoff frequency (6hr)
N  = 5               # the order of the butterworth filter
prepf = butter_filter(prep, N, fs, fc, btype='high')
tempf = butter_filter(temp, N, fs, fc, btype='high')
```

```python
# Settings for plot
plt.rcParams["axes.labelsize"] = 15
plt.rcParams["axes.titlesize"] = 15
plt.rcParams["xtick.labelsize"] = 15
plt.rcParams["ytick.labelsize"] = 15
```

### 1. Test on temperature and relative humidity -- minute data

```python
lagmax, L = 300, 1000
# Analysis on the original data
lagset, rhoxmpy, rhoympx = extended_ccm(temp[:L], prep[:L], temp[:L], prep[:L], 2, lagmax, 1)
lagset, rhoympz, rhozmpy = extended_ccm(prep[:L], temp[:L], prep[:L], temp[:L], 2, lagmax, 1)
lagset, mirxy, mixy = conductSST_withlag(temp[:L], prep[:L], lagmax, returnTrue=True, approach='kde_cuda')
rhoxy = corrcoefs(temp[:L], prep[:L], lagset)
# Analysis on the filtered data
lagset, rhoxmpyf, rhoympxf = extended_ccm(tempf[:L], prepf[:L], tempf[:L], prepf[:L], 2, lagmax, 1)
lagset, rhoympzf, rhozmpyf = extended_ccm(prepf[:L], tempf[:L], prepf[:L], tempf[:L], 2, lagmax, 1)
lagset, mirxyf, mixyf = conductSST_withlag(tempf[:L], prepf[:L], lagmax, returnTrue=True, approach='kde_cuda')
rhoxyf = corrcoefs(tempf[:L], prepf[:L], lagset)
```

```python
plot_extended_ccm_xy_withmi(lagset, rhoxmpy, rhoympx, rhoxy, mirxy, mixy,
                            rhoxmpyf, rhoympxf, rhoxyf, mirxyf, mixyf)
plt.show()
```

### 2. Tests on the accumulated and averaged data -- hourly data

```python
# Define
lagmax, L = 30, 1000
# Hourly data
interval = 60
prep_hr  = aggregate(prep, interval, method='accumulate')
temp_hr  = aggregate(temp, interval, method='average')
# Hourly data (filtered)
prepf_hr  = aggregate(prepf, interval, method='accumulate')
tempf_hr  = aggregate(tempf, interval, method='average')
```

```python
lagmax, s, L = 30, 10000, 1000
# Conduct CCM on precip and temp
lagset, rhoxmpy, rhoympx = extended_ccm(temp_hr[s:s+L], prep_hr[s:s+L], temp_hr[s:s+L], prep_hr[s:s+L], 2, lagmax, 1)
lagset, rhoympz, rhozmpy = extended_ccm(prep_hr[s:s+L], temp_hr[s:s+L], prep_hr[s:s+L], temp_hr[s:s+L], 2, lagmax, 1)
lagset, mirxy, mixy = conductSST_withlag(temp_hr[s:s+L], prep_hr[s:s+L], lagmax, returnTrue=True, approach='kde_cuda')
rhoxy = corrcoefs(temp_hr[s:s+L], prep_hr[s:s+L], lagset)
# Analysis on the filtered data
lagset, rhoxmpyf, rhoympxf = extended_ccm(tempf_hr[s:s+L], prepf_hr[s:s+L], tempf_hr[s:s+L], prepf_hr[s:s+L], 2, lagmax, 1)
lagset, rhoympzf, rhozmpyf = extended_ccm(prepf_hr[s:s+L], tempf_hr[s:s+L], prepf_hr[s:s+L], tempf_hr[s:s+L], 2, lagmax, 1)
lagset, mirxyf, mixyf = conductSST_withlag(tempf_hr[s:s+L], prepf_hr[s:s+L], lagmax, returnTrue=True, approach='kde_cuda')
rhoxyf = corrcoefs(tempf_hr[s:s+L], prepf_hr[s:s+L], lagset)
```

```python
# Plots
plot_extended_ccm_xy_withmi(lagset, rhoxmpy, rhoympx, rhoxy, mirxy, mixy,
                            rhoxmpyf, rhoympxf, rhoxyf, mirxyf, mixyf)
plt.show()
```

### 2.2. Tests on the accumulated and averaged data -- daily data

```python
# Daily data
interval = 60*24
prep_d  = aggregate(prep, interval, method='accumulate')
temp_d  = aggregate(temp, interval, method='average')
# Daily data (filtered)
prepf_d  = aggregate(prepf, interval, method='accumulate')
tempf_d  = aggregate(tempf, interval, method='average')
```

```python
# Conduct CCM on prep and temp
L, lagmax = 1000, 30
E, tau = 5, 6
lagset, rhoxmpy, rhoympx = extended_ccm(temp_d, prep_d, temp_d, prep_d, E, lagmax, tau)
lagset, rhoympz, rhozmpy = extended_ccm(prep_d, temp_d, prep_d, temp_d, E, lagmax, tau)
lagset, mirxy, mixy = conductSST_withlag(temp_d, prep_d, lagmax, returnTrue=True, approach='kde_cuda')
rhoxy = corrcoefs(temp_d, prep_d, lagset)
# Analysis on the filtered data
lagset, rhoxmpyf, rhoympxf = extended_ccm(tempf_d, prepf_d, tempf_d, prepf_d, E, lagmax, tau)
lagset, rhoympzf, rhozmpyf = extended_ccm(prepf_d, tempf_d, prepf_d, tempf_d, E, lagmax, tau)
lagset, mirxyf, mixyf = conductSST_withlag(tempf_d, prepf_d, lagmax, returnTrue=True, approach='kde_cuda')
rhoxyf = corrcoefs(tempf_d, prepf_d, lagset)
```

```python
# Plots
plot_extended_ccm_xy_withmi(lagset, rhoxmpy, rhoympx, rhoxy, mirxy, mixy,
                            rhoxmpyf, rhoympxf, rhoxyf, mirxyf, mixyf)
plt.show()
```

```python
# Plotting function
def plot_extended_ccm_xy_withmi(lagset, rhoxmpy, rhoympx, rhoxy, mirxy, mixy,
                                rhofxmpy, rhofympx, rhofxy, mirfxy, mifxy):
    """Plot the correlation coefficients and the CCM skills in terms of different lags for 2 nodes."""
    # Get the lags with the maximum skill
    maxympx, maxxmpy = lagset[np.argmax(rhoympx)], lagset[np.argmax(rhoxmpy)]
    maxfympx, maxfxmpy = lagset[np.argmax(rhofympx)], lagset[np.argmax(rhofxmpy)]

    # Get the lags with the correlation coefficients
    maxrhoxy = lagset[np.argmax(rhoxy)]
    maxrhofxy = lagset[np.argmax(rhofxy)]

    # Get the lags with the maximum skill
    maxmixy = lagset[np.argmax(mixy)]
    maxmifxy = lagset[np.argmax(mifxy)]
    
    xyt = mirxy==True
    xyf = mirxy==False
    xyft = mirfxy==True
    xyff = mirfxy==False

    # Plot
    plt.rcParams["figure.figsize"] = (16, 10)
    gs = gridspec.GridSpec(3, 2)
    gs.update(wspace=0.4, hspace=0.4)

    ax = plt.subplot(gs[0, 0])
    ax.plot(lagset, rhoxmpy, 'b', label='x xmp y')
    ax.plot(lagset, rhoympx, 'r', label='y xmp x')
    ax.set_xlabel('Lag')
    ax.set_ylabel('Correlation Coefficient')
    ax.set_ylim([-.5, 1.1])
    ax.set_title('Max. lag (x xmp y): %d; Max. lag(y xmp x): %d.' % (maxxmpy, maxympx))
    ax.legend(loc='lower right')

    ax = plt.subplot(gs[1, 0])
    ax.scatter(lagset[xyt], np.ones(xyt.sum()), color='red', label='MI SST True')
    ax.scatter(lagset[xyf], np.ones(xyf.sum()), color='blue', label='MI SST False')
    ax.plot(lagset, mixy, label='I value')
    ax.set_xlabel('Lag')
    ax.set_ylabel('I[x(t), y(t+Lag)]')
    ax.set_title('Max. lag: %d;' % maxmixy)
    ax.legend(loc='lower right')

    ax = plt.subplot(gs[2, 0])
    ax.plot(lagset, rhoxy, 'k', label='rho(x, y)')
    ax.set_xlabel('Lag')
    ax.set_ylabel('Rho[x(t), y(t+Lag)]')
    ax.set_title('Max. lag: %d;' % maxrhoxy)
    ax.set_ylim([-1.1, 1.1])
    
    ax = plt.subplot(gs[0, 1])
    ax.plot(lagset, rhofxmpy, 'b', label='x xmp y')
    ax.plot(lagset, rhofympx, 'r', label='y xmp x')
    ax.set_xlabel('Lag')
    ax.set_ylabel('Correlation Coefficient')
    ax.set_ylim([-.5, 1.1])
    ax.set_title('Max. lag (x xmp y): %d; Max. lag(y xmp x): %d.' % (maxfxmpy, maxfympx))
    ax.legend(loc='lower right')

    ax = plt.subplot(gs[1, 1])
    ax.scatter(lagset[xyft], np.ones(xyft.sum()), color='red', label='MI SST True')
    ax.scatter(lagset[xyff], np.ones(xyff.sum()), color='blue', label='MI SST False')
    ax.plot(lagset, mifxy, label='I value')
    ax.set_xlabel('Lag')
    ax.set_ylabel('I[x(t), y(t+Lag)]')
    ax.set_title('Max. lag: %d;' % maxmifxy)
    ax.legend(loc='lower right')

    ax = plt.subplot(gs[2, 1])
    ax.plot(lagset, rhofxy, 'k', label='rho(x, y)')
    ax.set_xlabel('Lag')
    ax.set_ylabel('Rho[x(t), y(t+Lag)]')
    ax.set_title('Max. lag: %d;' % maxrhofxy)
    ax.set_ylim([-1.1, 1.1])
```
