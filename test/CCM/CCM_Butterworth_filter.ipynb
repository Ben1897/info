```python
# Read the Data
filepath = '../data/SFP2_AllData.mat'
times, data, header, doy = parse_SFP(filepath)
precip, rh, temp = data[:, 1], data[:, 2], data[:, 3]
solar, wind      = data[:, 4], data[:, 5]
dsize = temp.size
```

```python
# Interpolate the data for NaNs
precip, rh = interpolate_nan(precip), interpolate_nan(rh)
temp, solar = interpolate_nan(temp), interpolate_nan(solar)
wind = interpolate_nan(wind)
```

## Filter frequencies with periods larger than 6hrs by using the Butterworth filter

```python
fs = 1./60.          # the sample rate
fc = 1./(60.*60.*6.) # the cutoff frequency (6hr)
wn = fc/(.5*fs)      # the relative cutoff frequency
N = 5                # the order of the filter
b, a = butter(N, wn, btype='high')   # should 'high' be here for bandpass?
tempf = lfilter(b, a, temp)
```

### Plot the original and the filtered data

```python
gs = gridspec.GridSpec(1, 1)
ax = plt.subplot(gs[0, 0])
ax.plot(temp, 'r', label='original')
ax.plot(tempf, 'b', label='filtered')
ax.set_xlabel('Time')
ax.set_ylabel('Temp')
ax.legend(loc='lower right')
plt.show()
```

### Calculate the frequencies of the original and filtered data and plot them

```python
itemp = fft(temp)[1:dsize/2]
itempf = fft(tempf)[1:dsize/2]
freq=fftfreq(temp.size, d=60)[1:dsize/2]
```

```python
gs = gridspec.GridSpec(1, 1)
ax = plt.subplot(gs[0, 0])
ax.loglog(freq, np.abs(itemp), 'r', label='original')
ax.loglog(freq, np.abs(itempf), 'b', label='filtered')
ax.axvline(x=fc, color='k', linewidth=2, linestyle='--', label='cutoff')
ax.set_xlabel('frequency (Hz)')
ax.set_ylabel('Power')
ax.legend()
plt.show('')
```
